```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the coefficients of the linear equation ax + by + cz = d
a, b, c, d = 2, 3, -1, 6

# Create a grid of x and y values
x = np.linspace(-10, 10, 100)  # X range
y = np.linspace(-10, 10, 100)  # Y range
x, y = np.meshgrid(x, y)       # Create a meshgrid for x and y

# Calculate z based on the equation z = (d - ax - by) / c
z = (d - a*x - b*y) / c

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, alpha=0.7, cmap='viridis')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot of the Plane')

# Show the plot
plt.show()
```
